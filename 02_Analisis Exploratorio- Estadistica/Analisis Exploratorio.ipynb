Vamos a buscar si tenemos una dependencia entre el precio de una casa y las variables que creemos puedan predecir el valor de la casa. 

## Precio de venta vs Area_habitable

```python
# Columna 

var = "GrLivArea"
```

```python
# Concatenar las dos columnas que queremos comparar.
data = pd.concat([df_train["SalePrice"],df_train[var]], axis= 1)
```

```python
# Veamos ahora las primeras filas
data.head()
```

```python
data.plot.scatter(x = var, y = "SalePrice", ylim=(0,800000))
```

Se puede observar que hay una relacion positiva. Es decir, sospechamos de una realción lineal entre estas dos variables. 

## Precio de Venta vs Sotano

```python
var = "TotalBsmtSF"
data = pd.concat([df_train["SalePrice"],df_train[var]], axis = 1)
data.plot.scatter(x = var, y ="SalePrice", ylim=(0,800000))
```

## Precio Venta vs Calidad


Relaciones categóricas, es decir el precio de la casa con las variables categoricas. 


```python
var = "OverallQual"
```

```python
# Concatenar las columnas OverrallQual con el precio de venta. 
data = pd.concat([df_train["SalePrice"],df_train[var]], axis = 1)
data.head()
```

```python
# Vamos hacer un gráfico en dos variables.
f, ax = plt.subplots(figsize = (10,10)) # subplot
fig = sns.boxplot(x = var, y = "SalePrice", data = data) # gráfico de caja
fig.axis(ymin=0,ymax=800000) # los parametro del eje y
```

Se nota una relación entre la calidad y el precio de la casa. 

## Precio de venta vs Construccion

```python
# Escoger la variable categorica
var = "YearBuilt"
# Concatenar las columnas OverrallQual con el precio de venta. 
data = pd.concat([df_train["SalePrice"],df_train[var]], axis = 1)
data.head()

```

```python
# Armar la grafica
f, ax = plt.subplots(figsize = (20,8)) # subplot
fig = sns.boxplot(x = var, y = "SalePrice", data = data) # gráfico de caja
fig.axis(ymin=0,ymax=800000) # los parametro del eje y
plt.xticks(rotation = 90) 
```

No es una tendencia muy fuerte pero las casas mas nuevas son mas caras además que se nota que es mas importante la calidad que el año de construcción.

# Matriz de correlación 
Nos interesan para ver que variables estan correlacionadas con el precio de venta

```python
corrmat = df_train.corr() 
corrmat
```

Sin embargo la informacion se aprecia mejor de manera gráfica por medio de un mapa de calor. 

```python
# Headmap
f, ax = plt.subplots(figsize = (12,9)) # subplot
sns.heatmap(corrmat, vmax= .8, square = True) # vmax es el gradiente. 
```

En este caso los cuadros blancos son los que indican mayor correlación. Vamos a seleccionar las variables que mas nos deberian interesar.   

```python
cols = corrmat.nlargest(10, "SalePrice")["SalePrice"].index
cols # Las 10 variables mas correlacionas. 
```

```python
cm= np.corrcoef(df_train[cols].values.T)
cm # Arreglo de correlacion transpuesto.
```

```python
sns.set(font_scale = 1.25)
```

```python
hm = sns.heatmap(cm, cbar = True, annot = True, square = True, fmt = ".2f", annot_kws = {"size":10,}, yticklabels = cols.values, xticklabels = cols.values)
plt.show()
```
